```python
#Importing Libraries
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler, LabelEncoder
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score
```

```python
#Load the data
customers = pd.read_csv('Mall_Customers.csv')
```

```python
numeric_columns = customers.select_dtypes(include=[np.number]).columns
print(numeric_columns)
```

```python
non_numeric_columns = customers.select_dtypes(exclude=[np.number]).columns
print(non_numeric_columns)
```

```python
#Null value Numeric columns filling with mean values
customers[numeric_columns] = customers[numeric_columns].fillna(customers[numeric_columns].mean())
print(customers[numeric_columns])
```

```python
#Converting non numeric columns to numeric columns
label_encoder = LabelEncoder()
```

```python
for column in non_numeric_columns:
    customers[column] = label_encoder.fit_transform(customers[column].astype(str))
```

```python
#Selecting Features
features = ['Age', 'Annual Income (k$)', 'Spending Score (1-100)']
X = customers[features]
```

```python
#Feature Scaling
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

sse = []
for k in range(1,11):
    kmeans = KMeans(n_clusters = k, random_state = 43)
    kmeans.fit(X_scaled)
    sse.append(kmeans.inertia_)
```

```python
plt.figure(figsize=(20, 12))
plt.plot(range(1, 11), sse, marker = 'o')
plt.xlabel('No. of Clusters')
plt.ylabel('StandardScaler')
plt.title('Elbow Method')
plt.show()
```

```python
silhouette_scores = []
for k in range(2,11):
    kmeans = KMeans(n_clusters=k, random_state = 43)
    kmeans.fit(X_scaled)
    silhouette_scores.append(silhouette_score(X_scaled, kmeans.labels_))
```

```python
plt.figure(figsize=(20, 12))
plt.plot(range(2, 11), silhouette_scores, marker = 'o')
plt.xlabel('No. of Clusters')
plt.ylabel('Silhouette Score')
plt.title('Silhouette Score Method')
plt.show()
```

```python
clusters_k = 5
kmeans = KMeans(n_clusters=clusters_k, random_state = 43)
customers['Clusters'] = kmeans.fit_predict(X_scaled)
```

```python
plt.figure(figsize=(20, 12))
sns.scatterplot(data=customers, x='Annual Income (k$)', y = 'Spending Score (1-100)', hue = 'Clusters', palette = 'coolwarm', s = 200)
plt.xlabel('Annual Income (k$)')
plt.ylabel('Spending Score (1-100)')
plt.title('Variety of Customers')
plt.legend()
plt.show()
```

```python
cluster_centres = scaler.inverse_transform(kmeans.cluster_centers_)
cluster_centres_df = pd.DataFrame(cluster_centres, columns = features )
cluster_centres_df['Clusters'] = range(clusters_k)
```

```python
plt.figure(figsize=(20, 12))
sns.heatmap(cluster_centres_df.set_index('Clusters'), annot=True, cmap='coolwarm')
plt.title('Cluster centers')
plt.show()
```
